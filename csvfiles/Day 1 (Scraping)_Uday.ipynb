### 4.Function to to scrape 1st page

```python
import pandas as pd
def quotes(soup):
    author_name = [i.text.strip() for i in soup.find_all('span',class_='authorOrTitle')]
    quotes = [list(i)[0].strip() for i in soup.find_all('div',class_='quoteText')]
    likes = [int(i.text.strip().split()[0]) for i in soup.find_all('a',class_='smallText')]
    return pd.DataFrame({"Authors":author_name,'Quotes':quotes,"Likes":likes})
```

```python
df = quotes(soup)
df
```

```python
df.shape
```

```python
def quotes_funny(soup1):
    author_name = [i.text.strip() for i in soup1.find_all('span',class_='authorOrTitle')]
    quotes = [list(i)[0].strip() for i in soup1.find_all('div',class_='quoteText')]
    likes = [int(i.text.strip().split()[0]) for i in soup1.find_all('a',class_='smallText')]
    tags = [i.text.split()[1:] for i in [i for i in soup1.find_all('div',class_='greyText smallText left')]]
    return pd.DataFrame({"Authors":author_name,'Quotes':quotes,"Likes":likes,"tags":tags})
```

## Motivational Quotes

```python
def quotes_mot(soup1):
    author_name = [i.text.strip() for i in soup1.find_all('span',class_='authorOrTitle')]
    quotes = [list(i)[0].strip() for i in soup1.find_all('div',class_='quoteText')]
    likes = [int(i.text.strip().split()[0]) for i in soup1.find_all('a',class_='smallText')]
    tags = [i.text.split()[1:] for i in [i for i in soup1.find_all('div',class_='greyText smallText left')]]
    return pd.DataFrame({"Authors":author_name,'Quotes':quotes,"Likes":likes,"tags":tags})
```

```python
stud3 = 'https://www.goodreads.com/quotes/tag/motivational?page={}'
```

```python
tables3=[]
from tqdm import tqdm
import time
for i in tqdm(range(1,11)):
    url_all = stud3.format(i)
    html_all = requests.get(url_all).text
    time.sleep(5)
    soup3 = BeautifulSoup(html_all,'html.parser')
    tables3.append(quotes_mot(soup3))
```
